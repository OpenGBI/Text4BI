##### 1.计算智能洞察中trend的数据

```python
# Assuming 'data.json' is your JSON file located in the current directory.
# Let's read the JSON file and convert it into a dictionary.

import json

# Read JSON data into the datastore variable
if __name__ == '__main__':
    with open('data/trendDataRaw2.json', 'r') as f:
        json_data = json.load(f)

    # Print the type to confirm it's a dictionary
    print(type(json_data))
    # Print the content
    print(json.dumps(json_data, indent=2))

```

```python
import json
from datetime import datetime

# Function to convert date string into a simplified format
def simplify_date(date_str):
    # Extract just the date part before the "~"
    date_part = date_str.split('~')[0]
    # Convert to datetime object and then format as required
    return datetime.strptime(date_part, "%Y-%m-%d %H:%M:%S").strftime("%Y-%m-%d")

# New JSON structure
new_json = {"data": [], "predictData": [] , "tagData": []}

# Loop through each entry to transform and add to the new structure
for entry in json_data['data']:
    for key, values in entry.items():
        if key.startswith("D"):
            if values[0]['actualValue'] is None:
                actual_value = 'none'
                predicted_value = values[0]['predictValue']
            else:
                actual_value = values[0]['actualValue']
                predicted_value = 'none'
            standard_value = values[0]['standardValue']
        elif key == "groupKey":
            date_value = list(values.values())[0]
            simplified_date = simplify_date(date_value)
            # Add to 'data' list
            new_json['data'].append({
                "date": simplified_date,
                "value1": actual_value,
                "category": "real" # Assuming all categories are 'real' for this example
            })
            new_json['predictData'].append({
                "date": simplified_date,
                "value1": predicted_value,
                "category": "predict" # Assuming all categories are 'real' for this example
            })
            # Add to 'tagData' list
            new_json['tagData'].append({
                "date": simplified_date,
                "value2": standard_value,
                "category": "regression" # Assuming all categories are 'real' for this example
            })

# Displaying the transformed JSON
print(json.dumps(new_json, indent=2))

with open('output/trendData2.json', 'w') as f:
    json.dump(new_json, f)
```

```python
# Assuming 'data.json' is your JSON file located in the current directory.
# Let's read the JSON file and convert it into a dictionary.

import json

# Read JSON data into the datastore variable
if __name__ == '__main__':
    with open('data/abnormalDataRaw2.json', 'r') as f:
        json_data = json.load(f)

    # Print the type to confirm it's a dictionary
    print(type(json_data))
    # Print the content
    print(json.dumps(json_data, indent=2))
```

```python
import json
from datetime import datetime

# Function to convert date string into a simplified format
def simplify_date(date_str):
    # Extract just the date part before the "~"
    date_part = date_str.split('~')[0]
    # Convert to datetime object and then format as required
    return datetime.strptime(date_part, "%Y-%m-%d %H:%M:%S").strftime("%Y-%m-%d")

# New JSON structure
new_json = {"data": []}

# Loop through each entry to transform and add to the new structure
for entry in json_data['data']:
    for key, values in entry.items():
        if key.startswith("D"):
            # value = values[0]['actualValue']
            # min = value[0]['standardValueLower']
            # predict = value[o]['standardValueMiddle']
            # max = value[0]['standardValueUpper'
            value = values[0]
            actual_value = value['actualValue']  # 这是一个浮点数
            min = value.get('standardValueLower')  # 这里使用 .get() 以避免 KeyError
            predict = value.get('standardValueMiddle')
            max = value.get('standardValueUpper')
            if actual_value > max:
                category = "upperAbnormal"
            elif actual_value < min:
                category = "lowerAbnormal"
            else:
                category = "normal"
        elif key == "groupKey":
            date_value = list(values.values())[0]
            simplified_date = simplify_date(date_value)
            # Add to 'data' list
            new_json['data'].append({
                "category": category,
                "date": simplified_date,
                "min": min,
                "max": max,
                "predict": predict,
                "value": actual_value,
            })

# Displaying the transformed JSON
print(json.dumps(new_json, indent=2))

with open('output/abnormalData2.json', 'w') as f:
    json.dump(new_json, f)
```
